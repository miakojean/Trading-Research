```python
import pandas as pd
```

```python
df = pd.read_excel("./Data/GER40CashData.xlsx")
df.head(5)
```

```python
# Supprimer les lignes avec valeurs manquantes
df.dropna(inplace=True)
```

```python
# Convertir la colonne Time en datetime
df['Time'] = pd.to_datetime(df['Time'])
```

```python
df.head(5)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(df['Time'], df['Close'], label='Prix')
plt.plot(df['Time'], df['Ichimoku_Tenkan'], label='Tenkan-Sen')
plt.plot(df['Time'], df['RSI_14'], label='RSI 14', alpha=0.5)
plt.legend()
plt.show()
```
